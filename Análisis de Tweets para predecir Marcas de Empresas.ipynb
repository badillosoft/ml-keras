## Generar el Corpus de Aprendizaje

En nuestro an√°lisis vamos a entrenar una RNN para que aprenda a reconocer un tweet como parte de una empresa. Es decir, dado un tweet (o texto en general) queremos predecir que marca se adapta mejor al tweet.

Por lo tanto, primero generaremos un corpus, el cu√°l especifique que tweet pertenece a que marca y as√≠ poder entrenar a la red. Una vez generado el corpus tendremos que predecir para un tweet arbitrario cu√°l es la marca que m√°s se asocia al tweet.

Vamos entonces a partir generando una lista de marcas, d√≥nde se indique la etiqueta de la marca, su `screen_name` de twitter y el n√∫mero m√°ximo de ejemplos que tomar√° de twitter.

```python
marcas = [
    {
        "label": "nike",
        "feed": "nikemexico",
        "count": 10000
    },
    {
        "label": "coca-cola",
        "feed": "cocacolamx",
        "count": 10000
    },
    {
        "label": "dominos",
        "feed": "dominosmx",
        "count": 10000
    }
]
```

Vemos a definir una funci√≥n que nos devuelva los tweets en el formato de lista de diccionarios.

Ejemplo: `[{ "label": "nike", "text": "..." }, { "label": "nike", "text": "..." }, ...]`

```python
import tweepy

def get_tweets(api, label, feed, count):
    tweets = []
    for tweet in tweepy.Cursor(api.user_timeline, timeout=1000, screen_name=feed).items(count):
        if tweet.in_reply_to_status_id == None:
            tweets.append({
                "id": tweet.id,
                "label": label,
                "text": tweet.text,
                "date": tweet.created_at
            })
    print("Se encontraron {} tweets de {}".format(len(tweets), feed))
    return tweets

print(get_tweets(api, "dominos", "dominosmx", 100))
```

Vamos ahora recolectar los tweets para todas las marcas.

```python
tweets = []

for marca in marcas:
    tweets.extend(get_tweets(api, marca["label"], marca["feed"], marca["count"]))
    
print("Se encontraron {} tweets de todas las marcas".format(len(tweets)))
```

Ahora que recolectamos todos los tweets para todas las marcas y ya est√°n etiquetados, procederemos a guardarlos en un corpus (un `DataFrame` de pandas).

```python
import pandas as pd

corpus = pd.DataFrame(tweets)

print(corpus.info())

print(corpus.head())
```

Finalmente guardamos el corpus con los tweets etiquetados para cada marca, para poder utilizarlo en an√°lisis posteriores.

```python
corpus.to_csv(open("tweet_corpus.csv", "w"), index=False, encoding="utf-8")

print("Corpus almacenado")
```

## Clasificador de Texto a Marca

Vamos a contruir un clasificador RNN para predecir una Marca de Empresa a partir de un texto (por ejemplo un tweet).

```python
import pandas as pd

corpus = pd.read_csv("http://badillosoft.com/tweet_corpus.csv")

corpus = corpus.sample(frac=1)

print(corpus.info())
```

```python
print(corpus.head())
```

```python
import text_util
from text_util import generate_sequences, generate_labels

x_train, x_test = generate_sequences(corpus["text"], thresh=1)

print(x_train[:3])
```

```python
y_train, y_test, categories = generate_labels(corpus["label"])

print(y_train[:10])
```

```python
print(categories)
```

```python
from keras.models import Sequential
from keras.layers import Embedding, GRU, LSTM, Dense

model = Sequential()

model.add(Embedding(
    input_dim=len(text_util.dictionary) + 1,
    output_dim=8,
    input_length=text_util.length
))

model.add(LSTM(16, return_sequences=True))
model.add(LSTM(8, return_sequences=True))
model.add(LSTM(4))

model.add(Dense(len(categories), activation="sigmoid"))

model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

print(model.summary())
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=32)
```

```python
metrics = model.evaluate(x_test, y_test)

print(metrics)
```
